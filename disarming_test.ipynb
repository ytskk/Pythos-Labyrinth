```python
import random


class Attributes:
    def __init__(
        self,
        dexterity: int,
        intelligence: int,
        luck: int = 40,
    ) -> None:
        self.dexterity = dexterity
        self.intelligence = intelligence
        self.luck = luck

    def __repr__(self) -> str:
        return f"{self.dexterity=}, {self.intelligence=}"


def generate_trap(level: int) -> Attributes:
    addition_intelligence = random.randint(1, level)
    addition_dexterity = level - addition_intelligence

    print(f"{addition_dexterity=}, {addition_intelligence=}")

    return Attributes(
        3 + addition_dexterity,
        4 + addition_intelligence,
    )


def calculate_trap_difficulty(
    trap: Attributes,
    hero_attributes: Attributes,
    hero_race_addition: int = 0,
) -> int:
    base_difficulty = 12

    hero_attributes = Attributes(
        hero_attributes.dexterity //10,
        hero_attributes.intelligence //10,
    )

    dexterity_difficulty = trap.dexterity - hero_attributes.dexterity
    intelligence_difficulty = trap.intelligence - hero_attributes.intelligence

    return (
        base_difficulty
        + dexterity_difficulty
        + intelligence_difficulty
        - hero_race_addition
    )
```

```python
human_attributes = Attributes(35, 35)
dwarf_attributes = Attributes(35, 40)
orc_attributes = Attributes(25, 25)
elf_attributes = Attributes(65, 55)

for level in range(1, 11):
    trap = generate_trap(level)
    print(f"Level {level} trap:")
    print(f"Difficulty for human is {calculate_trap_difficulty(trap, human_attributes)}")
    print(f"Difficulty for dwarf is {calculate_trap_difficulty(trap, dwarf_attributes, 4)}")
    print(f"Difficulty for orc is {calculate_trap_difficulty(trap, orc_attributes)}")
    print(f"Difficulty for elf is {calculate_trap_difficulty(trap, elf_attributes)}")
    print()
```
